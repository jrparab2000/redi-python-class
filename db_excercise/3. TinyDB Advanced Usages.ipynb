````python
from tinydb import TinyDB, Query
db = TinyDB('db-advanced.json')
City = Query()
````

## Use Tables

````python
germany = db.table('Germany')
````

````python
germany.all()
````

````python
# Get cities in Germnay that is within the state Bavaria
bavarian_cities = germany.search(City.state == "Bavaria")
bavarian_cities
````

````python
# Get the cities' English name
bavarian_city_names_en = [city["city_name_en"] for city in bavarian_cities]
print(bavarian_city_names_en)
````

````python
italy = db.table("Italy")
````

````python
italy.all()
````

````python
lombardian_cities = italy.search(City.region == "Lombardy")
lombardian_cities
````

## Advanced Queries

### With logical operations
Negate
```
~ (query)
```
Logical AND
```
(query1) & (query2)
```
Logical OR
```
(query1) & (query2)
```

````python
# Rewrite: Find the cities in Germany that are not in Bavaria
cities = germany.search(~ (City.state == "Bavaria"))
cities
````

### With customized test function for the database/table:
```
db.search(func, *args)
table.search(func, *args)
```
Example:
```
def condition(arg):
    return ...
    
Query.field.test(condition)
```

````python
# Find all the cities that has Umlaut (äöüÄÖÜ) in their German names
cities = germany.search(City.city_name_de.test(lambda name: any(umlaut in name for umlaut in "äöüÄÖÜ")))
cities
````

### With customized test function for the field with `Query()`:
```
Query().field.test(func, *args)
```
Example:
```
def condition(arg):
    return ...

table.search(condition)
```

````python
# Find the cities that are Stadtstaaten (i.e. city name equals to state name)
cities = germany.search(lambda city: city["city_name_en"] == city["state"])
cities
````

## Combine Advanced Functions with TinyDB

````python
# Get all the cities' English names from the table
cities = germany.all()
city_names = list(
    map(lambda x: x['city_name_en'], cities)
)
city_names
````

````python
# Get all the cities' English names for the cities in Bavaria
cities = germany.all()
city_names = list(
    map(lambda x: x['city_name_en'], 
        filter(lambda city: city['state'] == 'Bavaria', 
               cities)
       )
)
city_names
````
