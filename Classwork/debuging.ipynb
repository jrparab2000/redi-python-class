__Debugging__

using the logging library in the python this is a standard library.

```python
import logging
```

This is the verbosity levels: 

level set and higher will be printed all the things will not be printed so, if set to info then debug will not be printed

debug < info < warning < error < critical

```python
logger = logging.getlogger("my_logger")

logger.info("This is some informantion")
logger.waring("this is the warning")
logger.error("something is wrong")
```

```python
logging.basicConfig(level = logging.INFO)

logger.debug(" this is the lowest")
logger.info("This is some informantion")
logger.waring("this is the warning")
logger.error("something is wrong")
```

We can use a .log file to use it like below, this will create a log file named my_log.log this helps us to debug, important issues

```python
logger = logging.getLogger("my_logger")
logging.basicConfig(level=logging.DEBUG, filename = "my_log.log", format="%(asctime)s - %(name)s - %(levelname)s - %(message)s")
logger.debug(" this is the lowest")
logger.info("This is some informantion")
logger.warning("this is the warning")
logger.error("something is wrong")
```
