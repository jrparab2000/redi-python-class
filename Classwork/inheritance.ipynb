__Inharitance__

we can inherit other class which is called parent classs and other is child class. below is the parent class.

```python
class vehicle:
    def __init__(self, brand, max_speed):
        self.brand =brand
        self.max_speed = max_speed

    def info(self):
        print(f"This is a {self.brand} vechicle wit speed {self.max_speed}")
```

__Child__

below is the child class inheriting vehicle class. 

child ==> spaceship

parent ==> vechicle

```python
class spaceship(vehicle):
    pass

solar = spaceship("Spacex", 1000)
solar.info()

print(dir(solar))
```

__Method Overriding__

we can override the parent class methods by child class

```python
class Car(vehicle):
    def __init__(self, brand, max_speed, doors):
        super().__init__(brand, max_speed)
        self.doors = doors
    
    def info(self):
        print(f"This is a {self.brand} car with {self.doors} doors and speed {self.max_speed}")

    def engine(self):
        print("It has engines")

class Bike(vehicle):
    def __init__(self, brand, max_speed, bike_type):
        super().__init__(brand, max_speed)
        self.bike_type =bike_type

    def info(self):
        print(f"This is a {self.brand} bike with {self.bike_type} type and speed {self.max_speed}")
```

```python
car1 =Car('toyota', 180, 4)
car1.info()

bike = Bike('honda',150, 'sports')
bike.info()
```

__Dunder Methods__

Every class in python inherite from the object class by default

class Car: ----------> class Car(object):

This object class has basic methods this are call dunder methods like

`__init__`,`__str__`,`__eq__`, etc

```python
class Car:
    pass

print(dir(Car))
```

`__eq__`: will compare memory address of two objects not the attributes 
`__str__`: will retrun the string of the address 

```python
class Car1:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

car1 = Car1("Toyota","red")
car2 = Car1("Toyota","red")
car3 = car1

print("print car1: ", car1) #this will call __str__
print("car1 == car2: ", car1 == car2) #this will call __eq__
print("car1 == car3 ", car1 == car3)
```

__Overriding default dunder methods__

below we rewrite the default dunder methods, to our application.

```python
class Car2(Car1):
    def __str__(self):
        return f"{self.brand} and {self.color}"
    
    def __eq__(self, other):
        if not isinstance(other, Car2):
            return (self.brand == other.brand) and (self.color == other.color)
```

```python
car1 = Car2("Toyota","red")
car2 = Car2("Toyota","red")
car3 = car1

print("print car1: ", car1) #this will call __str__
print("car1 == car2: ", car1 == car2) #this will call __eq__
print("car1 == car3 ", car1 == car3)
```
