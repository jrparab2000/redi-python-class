__DataBase__

collection of organized data, can be accessed or modified.

__types__

MySQL, redis, SQLite, SOLServer, PostgreSQL, and mongoDB, etc. 

__Library__

`TinyDB`: CRUD => Create, Read, Update, and Delete. No server is used unlike SQL database. stores the data in a JSON file, making it easy to read and write. JSON => JavaScript Oject Notation. 

__Setup__

Installing ---> `pip install tinydb`



```python
from tinydb import TinyDB
```

__TinyDB__

below will create a file with the name provided if not presend. else it will just open the existing file

```python
db = TinyDB("user_db.json")
```

__CREATE__

`*.insert` function is used to add the data to the file, like shown below.

```python
db.insert({"name" : "Alice", "age" : 30, "city" : "new york"})
db.insert({"name" : "bob", "age" : 5, "city" : "austin"})
```

__READ__

`*.all` function is used to read the data from the json file.

```python
all_user = db.all()
print(all_user)
```

`Query` to search the specific data from the json file and return that only like below.

```python
from tinydb import Query

user = Query()
alice = db.search(user.name == "Alice")
print(alice)
```

__UPDATE__

using the `update` the exiting data in the JSON file using query and tinydb

```python
db.update({"age" : 26}, user.name == "Alice")
all_user = db.all()
print(all_user)
```

__DELETE__

`remove` function is used to delete the data or entry from the database.

```python
db.remove(user.name == "Alice")
all_user = db.all()
print(all_user)
```

__READ_ADVANCE__

```python
all_user = db.all()
all_bob = db.search(user.name == "bob")
print(all_bob)
first = db.get(user.name == "bob")
print(first)
print(first.doc_id)
```
